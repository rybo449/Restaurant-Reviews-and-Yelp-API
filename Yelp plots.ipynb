```python
import numpy as np
import csv
import time
import pandas as pd
import matplotlib.pyplot as plt
import sys
#%matplotlib inline
```

```python
f = open('cuisine_data_yelp.csv')
reader = csv.reader(f, delimiter = ',')
data = []
for row in reader:
    temp = []
    for i in row:
        i.strip()
        temp.append(i)
    data.append(temp)
df = pd.DataFrame(data)
```
